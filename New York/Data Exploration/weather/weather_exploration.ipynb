```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
temp_month = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/weather/data/temp_data_month_ny.csv', header=None)
crime_month = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/weather/data/crime_by_month.csv', header=None)
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
temp_series = pd.Series(temp_month[1].values, index=months)
crimes_series = pd.Series(crime_month[1].values, index=months)
month_temp = pd.DataFrame({'month':temp_series.index, 'value':temp_series.values})
month_crimes = pd.DataFrame({'month':crimes_series.index, 'value':crimes_series.values})
```

```python
figure,axis1 = plt.subplots()

axis1.plot(month_temp['value'], 'b-', label="Avg Temperature")
axis1.set_xlabel('Month')
axis1.set_ylabel('Avg Temperature', color='b')
axis1.tick_params('y', colors='b')

axis2 = axis1.twinx()
axis2.plot(month_crimes['value'], 'r-', label="Number of crimes")
axis2.set_ylabel('Number of crimes', color='r')
axis2.tick_params('y', colors='r')

axis1.tick_params(axis='x', rotation=90)
axis2.tick_params(axis='x', rotation=90)

plt.title('Temperature vs Crime rate')
plt.xticks(np.arange(12), months)
plt.show()
```
