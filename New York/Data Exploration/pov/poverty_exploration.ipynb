```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#crime by poverty per year
pov_bronx = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/bronx_poverty.csv', header=None)
pov_brooklyn = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/brooklyn_poverty.csv', header=None)
pov_manhattan = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/manhattan_poverty.csv', header=None)
pov_queens = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/queens_poverty.csv', header=None)
pov_staten = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/staten_poverty.csv', header=None)

#crime by year
crime_bronx = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/crime_by_year_bronx.csv', header=None)
crime_brooklyn = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/crime_by_year_brooklyn.csv', header=None)
crime_manhattan = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/crime_by_year_manhattan.csv', header=None)
crime_queen = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/crime_by_year_queens.csv', header=None)
crime_staten = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Exploration/pov/data/crime_by_year_staten.csv', header=None)
```

```python
#crime vs poverty in bronx
figure,axis1 = plt.subplots()

axis1.plot(pov_bronx[0], pov_bronx[1], 'b-', label="Poverty")
axis1.set_xlabel('year')
axis1.set_ylabel('Poverty rate', color='b')
axis1.tick_params('y', colors='b')

axis2 = axis1.twinx()
axis2.plot(crime_bronx[0], crime_bronx[1], 'r-', label="Number of crimes")
axis2.set_ylabel('Number of crimes', color='r')
axis2.tick_params('y', colors='r')

figure.tight_layout()
plt.title('Bronx: Number of crimes vs Poverty rate')
plt.show()
```
