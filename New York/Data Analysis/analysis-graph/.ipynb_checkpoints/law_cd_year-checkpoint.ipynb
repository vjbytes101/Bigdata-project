```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_fel = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/law_cd_year/crime_by_felony_year.csv', header=None)
data_misd = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/law_cd_year/crime_by_misdemeanor_year.csv', header=None)
data_vio = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/law_cd_year/crime_by_violation_year.csv', header=None)
```

```python
plt.plot(data_fel[0], data_fel[1], label="FELONY")
plt.plot(data_misd[0], data_misd[1], label="MISDEMEANOR")
plt.plot(data_vio[0], data_vio[1], label="VIOLATION")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel("Year")
plt.ylabel("Crimes")
plt.title("Level of Offense over the years")
plt.show()
```
