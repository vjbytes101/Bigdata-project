```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_j_fel = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/juris_law_cd/crime_by_juris_felony.csv', header=None)
data_j_misd = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/juris_law_cd/crime_by_juris_misdemeanor.csv', header=None)
data_j_vio = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/juris_law_cd/crime_by_juris_violation.csv', header=None)
```

```python
#crime by juris_felony
plt.bar(range(len(data_j_fel[0])), data_j_fel[2])
plt.xticks(range(len(data_j_fel[0])), data_j_fel[0], rotation=90)
plt.xlabel("Jurisdiction of Felony")
plt.ylabel("Number of Felonies")
plt.title("Felony by Jurisdiction")
plt.show()
```

```python
#crime by juris_misdemeanor
plt.bar(range(len(data_j_misd[0])), data_j_misd[2])
plt.xticks(range(len(data_j_misd[0])), data_j_misd[0], rotation=90)
plt.xlabel("Jurisdiction of Misdemeanor")
plt.ylabel("Number of Misdemeanor")
plt.title("Misdemeanor by Jurisdiction")
plt.show()
```

```python
#crime by juris_violation
plt.bar(range(len(data_j_misd[0])), data_j_misd[2])
plt.xticks(range(len(data_j_misd[0])), data_j_misd[0], rotation=90)
plt.xlabel("Jurisdiction of Violation")
plt.ylabel("Number of Violation")
plt.title("Violation by Jurisdiction")
plt.show()
```
