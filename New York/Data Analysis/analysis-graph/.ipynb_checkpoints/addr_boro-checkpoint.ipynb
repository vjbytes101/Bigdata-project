```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_bronx = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/addr_boro/crime_by_prec_bronx.csv', header=None)
data_brooklyn = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/addr_boro/crime_by_prec_brooklyn.csv', header=None)
data_manhattan = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/addr_boro/crime_by_prec_manhattan.csv', header=None)
data_queens = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/addr_boro/crime_by_prec_queens.csv', header=None)
data_staten = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/addr_boro/crime_by_prec_staten.csv', header=None)
```

```python
#crime precincts bronx
plt.plot(range(len(data_bronx[0])), data_bronx[1].astype(float), label="Bronx")
plt.xticks(range(len(data_bronx[0])), data_bronx[0])
plt.xlabel("Precincts")
plt.ylabel("Number of Crimes")
plt.title("Crimes by Precinct in Bronx")
plt.show()
```
