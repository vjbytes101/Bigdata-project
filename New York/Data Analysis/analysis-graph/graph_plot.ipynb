```python
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
queens_latlong = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/coordinates/queens_crimes_latlong.csv', header=None)
bronx_latlong = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/coordinates/bronx_crimes_latlong.csv', header=None)
brooklyn_latlong = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/coordinates/brooklyn_crimes_latlong.csv', header=None)
manhattan_latlong = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/coordinates/manhattan_crimes_latlong.csv', header=None)
statenisland_latlong = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/coordinates/staten_crimes_latlong.csv', header=None)
```

```python
left, bottom, right, top = -74.28, 40.49, -73.69, 40.93
m = Basemap(projection='cyl',llcrnrlat=bottom, urcrnrlat=top, llcrnrlon=left, urcrnrlon=right, lat_ts=bottom,resolution='i')
m.drawcoastlines()
m.fillcontinents(color='orange', lake_color='aqua')
x_1, y_1 = m(queens_latlong[1].values,queens_latlong[0].values)
x_2, y_2 = m(bronx_latlong[1].values,bronx_latlong[0].values)
x_3, y_3 = m(brooklyn_latlong[1].values,brooklyn_latlong[0].values)
x_4, y_4 = m(manhattan_latlong[1].values,manhattan_latlong[0].values)
x_5, y_5 = m(statenisland_latlong[1].values,statenisland_latlong[0].values)
m.scatter(x_1,y_1,marker='.',color='b',latlon=True,zorder=10)
m.scatter(x_2,y_2,marker='.',color='m',latlon=True,zorder=10)
m.scatter(x_3,y_3,marker='.',color='g',latlon=True,zorder=10)
m.scatter(x_4,y_4,marker='.',color='r',latlon=True,zorder=10)
m.scatter(x_5,y_5,marker='.',color='c',latlon=True,zorder=10)
plt.title("Scatter Plot of Crimes")
plt.show()
```
