```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/Vijay/Desktop/Sem2/big data project/Data Analysis/crime by day/crimebyday.csv', header=None)
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
data.set_index(0).loc[day_order].plot(kind = 'bar', color='red', legend = False)

plt.xlabel("Week Day")
plt.ylabel("Crimes")
plt.title("Crimes by Day in Week")
plt.show()
```
